```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# 2. Senior DS Check: Quick Audit
print(f"Shape: {df.shape}")
print("\n--- Data Types ---")
print(df.dtypes)

# 3. The "Dirty Data" Trap Check
# 'TotalCharges' is often read as object because of empty strings " "
# Let's force it to numeric and see how many fail
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"\nMissing values in TotalCharges after conversion: {df['TotalCharges'].isnull().sum()}")

# 4. Target Distribution (The most important metric for Day 1)
churn_dist = df['Churn'].value_counts(normalize=True) * 100
print("\n--- Churn Distribution ---")
print(churn_dist)

# Visualizing the Target
sns.countplot(x='Churn', data=df)
plt.title("Class Distribution (Imbalance Check)")
plt.show()
```
